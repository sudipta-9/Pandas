# Replace Function

```python
import pandas as pd
var = pd.read_csv("D:\\Projects\\Pandas\\Test.csv")
var
```

```python
var.replace(to_replace=1,value=22)
```

```python
var.replace(to_replace="Helsa",value="Ram")
```

```python
var.replace([1,2,3,4,5,6,7,8,9],22)
```

```python
var.replace("[A-Za-z]","python",regex=True)
```

```python
var.replace("[A-Z]","python",regex=True)
```

```python
var.replace({"firstname":'[A-Z]'},22,regex=True)
```

```python
var.replace(1,method="ffill")
```

```python
var.replace(1,method="bfill")
```

```python
var.replace(1,method="ffill",limit=2)
```

```python
var.replace(1,method="ffill",limit=2,inplace=True)
var
```
