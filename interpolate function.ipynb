# Interpolate Function

```python
import pandas as pd
var = pd.read_csv("D:\\Projects\\Pandas\\Test.csv")
var
```

```python
var.interpolate() #It's only work with digit not alphabet
```

```python
var.interpolate(method="linear")
```

```python
var.interpolate(method="linear",axis=0)
```

```python
var.interpolate(method="linear",axis=1) #All the datatypes need to be same type
```

```python
var.interpolate(limit=2) #Only fill the 2 data values
```

```python
var.interpolate(limit_direction="forward",limit=2) #Only fill the 2 data values forward direction
```

```python
var.interpolate(limit_direction="backward",limit=2) #Only fill the 2 data values backward direction
```

```python
var.interpolate(limit_direction="both",limit=2) #Only fill the 2 data values both direction
```

```python
var.interpolate(limir_area="inside") #It will fills all the null values inside the csv file
```

```python
var.interpolate(limir_area="outside") #It will remain null values in the csv file
```

```python
var.interpolate(limit_direction="both",limit=2,inplace=True) #Modify the original file and don't make copy
var 
```

```python
var.interpolate(limit_direction="both",limit=2,inplace=False) #Make the multiple copy and the original file reamin same
var 
```
