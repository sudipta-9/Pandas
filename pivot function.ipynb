# Pivot Function

```python
import pandas as pd
var = pd.DataFrame({"beng":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})
var
```

```python
var.pivot(index="beng",columns="st_name")
```

```python
var.pivot(index="beng",columns="st_name",values="eng") #Only show the values of eng
```

```python
var.pivot_table(index="beng",columns="st_name",aggfunc="mean")
```

```python
var.pivot_table(index="beng",columns="st_name",aggfunc="sum")
```

```python
var.pivot_table(index="beng",columns="st_name",aggfunc="sum",margins=True)
```
