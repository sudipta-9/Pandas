# Melt Function

```python
import pandas as pd
var = pd.DataFrame({"beng":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"math":[17,18,19,13,14,16]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["eng"]) #It show eng as index value
```

```python
pd.melt(var,id_vars=["eng"],var_name="python") #Change the variable name variable to python
```

```python
pd.melt(var,id_vars=["eng"],value_name="number") #Change the value name value to number
```
