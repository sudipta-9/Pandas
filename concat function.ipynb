# Concat Function

```python
import pandas as pd  
var1 = pd.Series([1,2,3,4])
var2 = pd.Series([11,21,31,41])  
pd.concat([var1, var2])  
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([var1, var2]) 
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([var1, var2],axis=0) #It will concat accross row wise
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([var1, var2],axis=1) #It will concat accross column wise
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([var1, var2],axis=1) #If data is not proper filled then it will show NAN
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([var1, var2],axis=1,join="outer") #It will show the full data values
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([var1, var2],axis=1,join="inner") #It will ommit the missing data 
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([var1, var2],axis=1,keys=["d1","d2"])
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([var1, var2],axis=0,keys=["d1","d2"])
```
