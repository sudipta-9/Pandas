# Dropna Function

```python
import pandas as pd
var = pd.read_csv("D:\\Projects\\Pandas\\Test.csv")
var
```

```python
var.dropna()
```

```python
var.dropna(axis = 0)
```

```python
var.dropna(thresh=1)
```

```python
var.dropna(thresh=2)
```
