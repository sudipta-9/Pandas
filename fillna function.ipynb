# Fillna Function

```python
import pandas as pd
var = pd.read_csv("D:\\Projects\\Pandas\\Test.csv")
var
```

```python
var.fillna("python")
```

```python
var.fillna({"firstname":"python","lastname":"c"})
```

```python
var.ffill() #Forward Fill
```

```python
var.bfill() #Backward Fill
```

```python
var.bfill(axis=0)
```

```python
var.bfill(axis=1)
```

```python
var.fillna(12,inplace=True)
var
```

```python
var.fillna("python",limit=2)
```
